```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/energydata_complete.csv")
df=df.drop(columns=['date','lights'])
df
```

```python
simple_linear_reg_df = df[['T2','T6']].sample(15,random_state=1)
simple_linear_reg_df
```

```python
sns.regplot(x="T2",y="T6",data=simple_linear_reg_df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df= pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
normalised_df
```

```python
feature_df= normalised_df.drop(columns=['T6','rv1','rv2'])
```

```python
T6=df['T6']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
x_train, x_test, y_train, y_test = train_test_split(feature_df,T6,test_size=0.4,random_state=1)
linear_model = LinearRegression()
linear_model.fit(x_test,y_test)
predicted_values=linear_model.predict(x_train)
rmse = np.sqrt(mean_squared_error(y_train,predicted_values))
round(rmse,3)
```

```python
from sklearn.metrics import mean_absolute_error
feature1_df=normalised_df.drop(columns=['Appliances'])
Appliances=df['Appliances']
x_train, x_test, y_train, y_test = train_test_split(feature1_df,Appliances,test_size=0.3,random_state=42)
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
predicted_values=linear_model.predict(x_test)
mae=mean_absolute_error(y_test,predicted_values)
round(mae,3)
```

```python
linear_model.fit(x_train,y_train)
predicted_values=linear_model.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test,predicted_values))
round(rmse,3)
```

```python
linear_model.fit(x_test,y_test)
predicted_values=linear_model.predict(x_train)
rmse = np.sqrt(mean_squared_error(y_train,predicted_values))
round(rmse,3)
```

```python
linear_model.fit(x_test,y_test)
predicted_values=linear_model.predict(x_train)
mae=mean_absolute_error(y_train,predicted_values)
round(mae,3)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=0.5)
ridge_reg.fit(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
# Extract the feature weights from the model and sort them
  weights = pd.Series(model.coef_, feat.columns).sort_values()
# Create a DataFrame to store the feature weights
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```
